```python
# import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class DescriptiveStatistics:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)
    
    def calculate_statistics(self):
        return self.data.describe()
    
    def get_variable_names(self):
        return self.data.columns.tolist()
    
    def get_variable_stats(self, variable_name):
        variable_data = self.data[variable_name]
        statistics = {
            'Media': variable_data.mean(),
            'Mediana': variable_data.median(),
            'Desviación estándar': variable_data.std(),
            'Q25': variable_data.quantile(0.25),
            'Q50': variable_data.quantile(0.50),
            'Q75': variable_data.quantile(0.75)
        }
        plt.hist(variable_data, bins='auto')
        plt.xlabel(variable_name)
        plt.ylabel('Frequencia')
        plt.title('Histograma de ' + variable_name)
        plt.show()
        return statistics
```

```python
# Ejemplo de uso
data_obj = DescriptiveStatistics("C:/Users/Daniel/Documents/Universidad Nacional/Diplomado IA/Bases/Iris.csv")
print("Variables disponibles:")
print(data_obj.get_variable_names())
print()
```

```python
variable_name = 'SepalLengthCm'  # Nombre de la variable a analizar
print("Estadísticas de la variable:", variable_name)
print(data_obj.get_variable_stats(variable_name))
```
